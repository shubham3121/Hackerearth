```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
train = pd.read_csv('train_data.csv')
test = pd.read_csv('test_data.csv')
```

```python
# Continuous variables
cont_cols = ['cont_1','cont_2','cont_3','cont_4','cont_5','cont_6','cont_7','cont_8','cont_9','cont_10','cont_11','cont_12',
            'cont_13','cont_14','cont_15','cont_16','cont_17','cont_18']

# Categorical variables
cat_cols = ['cat_1','cat_2','cat_3','cat_4','cat_5','cat_6','cat_7','cat_8','cat_9','cat_10','cat_11','cat_12','cat_13',
           'cat_14','cat_15','cat_16', 'cat_17','cat_18','cat_19','cat_20', 'cat_21', 'cat_22','cat_23']
```

```python
train.head()
```

```python
cat_cont = train[['cont_6','cont_11']].groupby('cont_11').agg(['mean'])

cat_cont.reset_index(inplace=True)

cat_cont.columns = list(map(''.join, cat_cont.columns.values))

cat_cont.columns = ['cont_11', 'cat_cont']

train = train.merge(cat_cont, on='cont_11')
```

```python
cat_cont = test[['cont_6','cont_11']].groupby('cont_11').agg(['mean'])

cat_cont.reset_index(inplace=True)

cat_cont.columns = list(map(''.join, cat_cont.columns.values))

cat_cont.columns = ['cont_11', 'cat_cont']

test = test.merge(cat_cont, on='cont_11')
```

```python
print "Number of categorical features:", len(cat_cols)
print 'Number of continuous features:', len(cont_cols)
```

```python
X = np.array(train.drop(['connection_id','cat_17','cont_1','target'], axis = 1))
y = train.target
```

```python
X_test = np.array(test.drop(['connection_id','cont_1','cat_17'], axis = 1))
```

```python
import lightgbm as lgb
import xgboost as xgb
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.cross_validation import KFold
from sklearn.cross_validation import cross_val_score
```

```python
def multAcc(pred, dtrain):
    label = dtrain.get_label()
    acc = accuracy_score(label, pred)
    return 'maccuracy', acc
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.15, random_state = 47)
```

```python
dtrain = xgb.DMatrix(data=X_train, label=y_train)
dvalid = xgb.DMatrix(data=X_valid, label=y_valid)
dtest = xgb.DMatrix(X_test)
```

```python
watchlist = [(dtrain, 'train'),(dvalid, 'valid')]
```

```python
params = {}
params['objective'] = 'multi:softmax'
params['eta'] = 0.01
params['max_depth'] = 15
params['silent'] = 0
params['max_bin'] = 256
params['max_leaf_nodes'] = 2**12
params['num_class'] = 3
params['colsample_bytree'] = 0.9
```

```python
xgb.cv(params=params, dtrain=dtrain, num_boost_round=300, nfold=5, feval=multAcc, seed=0, callbacks=[xgb.callback.print_evaluation(show_stdv=)])
```

```python
model = xgb.train(params, dtrain, 300, watchlist, maximize=True, verbose_eval=10, early_stopping_rounds=50, feval=multAcc)
```

```python
pred = model.predict(dtest)
```

```python
subm = pd.DataFrame()
```

```python
subm['connection_id'] = test.connection_id
subm['target'] = pred.astype('int')
subm.to_csv('submit.csv',index=False)
```
